```python
#ngrams function
def ngrams(s, N):
    ngram_list = []
    seq_len = len(s)
    num_ngrams = seq_len - (N-1)
    for start_ind in range(num_ngrams):
        ngram_list.append(tuple(s[start_ind:start_ind+N]))
    return(ngram_list)
```

```python
#make all lowercase and lemmatize
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(Articles):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(Articles)]

#make all lowercase
articles_clean['Articles']= articles_clean['Articles'][0:10].str.lower()
#lemmatizing
articles_clean['Articles_lemm'] = articles_clean.Articles[0:10].apply(lemmatize_text)
```

```python
#get bigrams and trigrams
from nltk.util import ngrams
from nltk import word_tokenize

articles_clean['bigram'] = articles_clean['Articles'][0:100].apply(lambda row: list(ngrams(word_tokenize(row), 2)))
articles_clean['trigram'] = articles_clean['Articles'][0:100].apply(lambda row: list(ngrams(word_tokenize(row), 3)))
articles_clean
```

## Test of lowercase, lemmatization, and ngrams with sample dataset

```python
#test w/ sample data set
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

df = pd.DataFrame(['Wow I love this', 'Cats have nine lives', 'Dogs are also great', 'plurals', 'circus canaries boss', 'ponies'], columns=['text'])
#lowercase
df['text']= df['text'].str.lower()
#lemmatizing
df['text_lemmatized'] = df.text.apply(lemmatize_text)
```
