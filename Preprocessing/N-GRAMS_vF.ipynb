```python
import pandas as pd
import numpy as np

articles_clean = pd.read_csv('new_data.csv',encoding='latin1')
articles_clean.head()
```

```python
#quick check
type(articles_clean['Articles'][0])
```

```python
#make all lowercase and lemmatize
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(Article):
    if not isinstance(Article, str) or len(Article)==0:
        return list('<blank>') 
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(Article)]

#make all lowercase
articles_clean['Articles']= articles_clean['Articles'].str.lower()
#lemmatizing
articles_clean['Articles_lemm'] = articles_clean.Articles.apply(lemmatize_text)
```
